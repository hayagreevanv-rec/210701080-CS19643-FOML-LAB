```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
data_X,data_y = load_iris(return_X_y=True,as_frame=True)
```

```python
print(data_X,data_y)
```

```python
sns.scatterplot(x='sepal length (cm)',y='petal length (cm)',data=data_X,hue=data_y)
plt.show()
```

```python
sns.catplot(x=data_y,y='sepal length (cm)',data=data_X ,kind='box',hue=data_y)
plt.show()
```

```python
sns.catplot(x=data_y,y='petal length (cm)',data=data_X ,kind='swarm',hue=data_y)
plt.show()
```

```python
X,y = load_iris(return_X_y=True)
```

```python
_,ax = plt.subplots()
```

```python
scatter =ax.scatter(X[:,0],X[:,1],c=y)
ax.set(xlabel=load_iris().feature_names[0],ylabel=load_iris().feature_names[1])
_ = ax.legend(scatter.legend_elements()[0],load_iris().target_names,loc='lower right',title='Classes')
```

```python
scatter =ax.scatter(X[:,1],X[:,2],c=y)
ax.set(xlabel=load_iris().feature_names[1],ylabel=load_iris().feature_names[2])
_ = ax.legend(scatter.legend_elements()[0],load_iris().target_names,loc='lower right',title='Classes')
```
